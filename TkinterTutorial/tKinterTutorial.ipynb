## First Step

## Buttons

```python
from tkinter import *
root = Tk()

# Create a Button Widget
# And positioning using the grid system (now they will stay always at the same relative position)
myButton1 = Button(root, text = "Click Me!").pack() 
myButton2 = Button(root, text = "Disabled", state = DISABLED).pack() # We can disable the button
myButton3 = Button(root, text = "I'm Bigger!!", padx = 10, pady = 0).pack() # We can disable the button
myButton5 = Button(root, text = "I'm Blue!", fg = "blue").pack() # We can change colors (Mac does not support change in buttons backgrounds)

# Adding functions to the Buttons
def myClick():
    myLabel = Label(root, text = "Look! I clicked a Button!")
    myLabel.pack()

myButton4 = Button(root, text = "Guess What?", command = myClick) # Function wothout parenthesis, it runs at all clicks
myButton4.pack() 

# Execute
root.mainloop()
```

## Inputs

```python
from tkinter import *
root = Tk()

# Create an Entry Widget
myEntry1 = Entry(root)
myEntry1.insert(0, "Enter your Name: ") # Write inside in box 0 
myEntry1.pack()
# myEntry2 = Entry(root, width = 50, bg = "blue", fg = "white").pack() # Blue box, white text
# myEntry3 = Entry(root, borderwidth = 5).pack() # Change border

# Saving entries with Entry.get()
def myClick():
    myLabel2 = Label(root, text = "Hello " + myEntry1.get() + "!")
    myLabel2.pack()

# Button gives name
myButton1 = Button(root, text = "Click Me", command = myClick) # Returns name
myButton1.pack() 

# Execute
root.mainloop()
```

## Basic Calculator

```python
from tkinter import *

# Window
root = Tk()
root.title("Calculator") # Title window
root.resizable(0,0) #by this command we can make the window not resizible

# Display
Display = Entry(root, width = 28, borderwidth = 5)
Display.grid(row=0, column=0, columnspan = 3) # It spans into 3 columns

#Define functions
def button_click(number): 
    Current = Display.get()
    Display.delete(0, END)
    Display.insert(0, Current + number) 

def button_clear():
    Display.delete(0,END)

def button_add():
    global Remember
    global Operation
    Operation = "+"
    Remember = Display.get()
    Display.delete(0,END)

def button_rest():
    global Remember
    global Operation
    Operation = "-"
    Remember = Display.get()
    Display.delete(0,END)

def button_mult():
    global Remember
    global Operation
    Operation = "*"
    Remember = Display.get()
    Display.delete(0,END)

def button_div():
    global Remember
    global Operation
    Operation = "/"
    Remember = Display.get()
    Display.delete(0,END)
    
def button_equal():
    global Remember
    global Operation
    if Operation == "+":
        Result = str(round(float(Remember) +  float(Display.get()),2))
    elif Operation == "-":
        Result = str(round(float(Remember) -  float(Display.get()),2))
    if Operation == "*":
        Result = str(round(float(Remember) *  float(Display.get()),2))
    if Operation == "/":
        Result = str(round(float(Remember) /  float(Display.get()),2))
    Display.delete(0,END)
    Display.insert(0,Result)

#Buttons
Button_1 = Button(root, text = "1", padx = 40, pady = 20, command=lambda: button_click("1"))
Button_2 = Button(root, text = "2", padx = 40, pady = 20, command=lambda: button_click("2"))
Button_3 = Button(root, text = "3", padx = 40, pady = 20, command=lambda: button_click("3"))
Button_4 = Button(root, text = "4", padx = 40, pady = 20, command=lambda: button_click("4"))
Button_5 = Button(root, text = "5", padx = 40, pady = 20, command=lambda: button_click("5"))
Button_6 = Button(root, text = "6", padx = 40, pady = 20, command=lambda: button_click("6"))
Button_7 = Button(root, text = "7", padx = 40, pady = 20, command=lambda: button_click("7"))
Button_8 = Button(root, text = "8", padx = 40, pady = 20, command=lambda: button_click("8"))
Button_9 = Button(root, text = "9", padx = 40, pady = 20, command=lambda: button_click("9"))
Button_0 = Button(root, text = "0", padx = 40, pady = 20, command=lambda: button_click("0"))
Button_float = Button(root, text = ".", padx = 42, pady = 20, command=lambda: button_click("."))
Button_add = Button(root, text = "+", padx = 40, pady = 20, command=button_add)
Button_rest = Button(root, text = "-", padx = 40, pady = 20, command=button_rest)
Button_mult = Button(root, text = "*", padx = 41, pady = 20, command=button_mult)
Button_div = Button(root, text = "/", padx = 41, pady = 20, command=button_div)
Button_equal = Button(root, text = "=", padx = 87, pady = 20, command=button_equal)
Button_clear = Button(root, text = "Clear", padx = 28, pady = 20, command= button_clear)

#ButtonsGrid
Button_1.grid(row = 3, column = 0)
Button_2.grid(row = 3, column = 1)
Button_3.grid(row = 3, column = 2)
Button_4.grid(row = 2, column = 0)
Button_5.grid(row = 2, column = 1)
Button_6.grid(row = 2, column = 2)
Button_7.grid(row = 1, column = 0)
Button_8.grid(row = 1, column = 1)
Button_9.grid(row = 1, column = 2)
Button_0.grid(row = 4, column = 0)
Button_add.grid(row = 4, column = 1)
Button_rest.grid(row = 4, column = 2)
Button_mult.grid(row = 5, column = 1)
Button_div.grid(row = 5, column = 2)
Button_float.grid(row = 5, column = 0)
Button_clear.grid(row = 6, column = 0)
Button_equal.grid(row = 6, column = 1, columnspan = 2)

# Execute
root.mainloop()
```

## Image Viewer App

```python
from tkinter import *
from PIL import ImageTk, Image 

# Window
root = Tk()
root.title("Photos")
root.resizable(0,0)

# Images
myImage1 = ImageTk.PhotoImage(Image.open("dice-1.png"))
myImage2 = ImageTk.PhotoImage(Image.open("dice-2.png"))
myImage3 = ImageTk.PhotoImage(Image.open("dice-3.png"))
myImage4 = ImageTk.PhotoImage(Image.open("dice-4.png"))
myImage5 = ImageTk.PhotoImage(Image.open("dice-5.png"))
myImage6 = ImageTk.PhotoImage(Image.open("dice-6.png"))
myImage7 = ImageTk.PhotoImage(Image.open("image_test1.jpg"))
Image_list = [myImage1, myImage2, myImage3, myImage4, myImage5, myImage6, myImage7]

# Grid
index = 0; # Auxiliar varible
myLabel = Label(root, image = Image_list[index])
myLabel.grid(row = 0, column = 0, columnspan = 3)

# Funct
def forward():

    global myLabel
    global ButtonNext
    global ButtonBack
    global index


    index += 1
    myLabel.grid_forget()
    myLabel = Label(root, image = Image_list[index])
    myLabel.grid(row = 0, column = 0, columnspan = 3)

    if index == (len(Image_list)-1):
        ButtonNext.configure(state = DISABLED)

    if index > 0:
        ButtonBack.configure(state = NORMAL)

def backward():

    global myLabel
    global ButtonNext
    global ButtonBack
    global index
    
    index -= 1
    myLabel.grid_forget()
    myLabel = Label(root, image = Image_list[index])
    myLabel.grid(row = 0, column = 0, columnspan = 3)

    if index < (len(Image_list)-1):
        ButtonNext.configure(state = NORMAL)

    if index == 0:
        ButtonBack.configure(state = DISABLED)

# Buttons
ButtonBack = Button(root, text = "Back", padx = 40, command = backward, state = DISABLED)
ButtonExit = Button(root, text = "Exit", padx = 40, command = root.quit)
ButtonNext = Button(root, text = "Next", padx = 40, command = forward)

ButtonBack.grid(row = 1, column = 0)
ButtonExit.grid(row = 1, column = 1)
ButtonNext.grid(row = 1, column = 2)

# Loop
root.mainloop()
```
